```python
import pandas as pd
from fuzzywuzzy import fuzz
from collections import defaultdict

# Path to the uploaded file

# Function to group column names based on their patterns
def group_column_names(column_names, similarity_threshold=85):
    groups = defaultdict(list)  # Dictionary to store grouped columns
    used_indices = set()  # To track already grouped columns
    
    for i, col in enumerate(column_names):
        if i in used_indices:  # Skip already grouped columns
            continue
        group_name = col  # First appearance of the column name becomes the group name
        groups[group_name].append((col, i))  # Add the column and its index to the group
        used_indices.add(i)  # Mark the index as used

        # Compare with remaining columns to find similar names
        for j, other_col in enumerate(column_names):
            if j != i and j not in used_indices:
                similarity = fuzz.partial_ratio(col.lower(), other_col.lower())
                if similarity >= similarity_threshold:
                    groups[group_name].append((other_col, j))
                    used_indices.add(j)  # Mark the index as used

    return groups

# Load the file and read column names
try:
    data = pd.read_csv(file_path, nrows=0)  # Read only the first row (header)
    column_names = data.columns.tolist()   # Extract column names as a list
    
    # Group column names
    grouped_columns = group_column_names(column_names)

    # Display the results as a table-like format
    print("Grouped Columns:")
    for group_name, columns in grouped_columns.items():
        print(f"Group Name: {group_name}")
        for col, idx in columns:
            print(f"  - Column: {col}, Original Index: {idx}")
        print()
except Exception as e:
    print(f"Error: {e}")
```
